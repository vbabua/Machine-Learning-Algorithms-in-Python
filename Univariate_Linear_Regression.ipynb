# Predicting Diabetes Progression Using Linear Regression


#Introduction to the Problem
In this example, we'll explore how we can use linear regression, to predict disease progression in diabetes patients based on their Body Mass Index (BMI). Linear regression is a method used to model the relationship between a dependent variable and one or more independent variables. Here, we'll focus on a simple linear regression model that uses just one independent variable.

# Dataset
We'll be using the Diabetes dataset provided by scikit-learn, a popular machine learning library in Python. This dataset is specifically designed for regression tasks and includes data on diabetes patients. It contains ten baseline variables: age, sex, body mass index (BMI), average blood pressure, and six blood serum measurements.

For our simple linear regression model, we'll use BMI as our independent variable (or feature) and disease progression one year after baseline as our dependent variable (or target). The choice of BMI is motivated by its known association with diabetes risk and progression, making it an insightful variable for prediction.

# Hypothesis Function
The hypothesis function for our linear regression model is given by:

$$
f_{wb}(x^{(i)}) = wx^{(i)} + b
$$

where
- \( f_{wb}(x^{(i)}) \) is the predicted value for the \(i\)-th example,
- \( w \) is the weight or coefficient,
- \( x^{(i)} \) is the feature value for the \(i\)-th example,
- \( b \) is the bias term.


# Compute cost Function


$$
J(w, b) = \frac{1}{2m} \sum_{i=1}^{m} \left( f_{wb}(x^{(i)}) - y^{(i)} \right)^2
$$

where:

- $$( J(w, b) ) $$is the cost function,
- $$( m )$$ is the number of training examples,
- $$( f_{wb}(x^{(i)}) )$$ is our hypothesis or prediction,
- $$( x^{(i)} )$$ is the input feature(s) of our \( i \)-th training example,
- $$( y^{(i)} )$$ is the actual output of our \( i \)-th training example,
- $$( w )$$ represents the model parameters (weights),
- $$( b )$$ represents the bias term.

```python
def compute_cost(x, y, w, b):
  m = x.shape[0]

  cost = 0

  for i in range(m):
    f = np.dot(w, x[i]) + b
    cost = (f - y[i]) ** 2
  cost /= (2 * m)

  return cost
```
